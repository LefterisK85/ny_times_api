### Collect the top headlines every 5 minutes for ~8 hours

```python
from time import sleep
import json, urllib, urllib.request
```

```python
def fetch_articles():
    """This function fetches the top 10 headlines from NY Times"""
    data = urllib.request.urlopen("https://api.nytimes.com/svc/search/v2/articlesearch.json?api-key=ca47399f4cfd4888bb95b6f4e5daef6f").read()
    data_json = json.loads(data)
    articles.extend(data_json['response']['docs'])
```

```python
articles = []
```

```python
# This loop will fetch the top 10 headlines every 5 minutes and will do that 100 times (approximatey for 8 hours)
for n in range(0,100):
    try:
        fetch_articles()
        sleep(300)
    except:
        print('Error after', len(articles)/10, 'samples')
```

```python
# The total number of articles collected over 8 hours
len(articles)
```

```python
# The description from the first 5 articles
snippets = []
for article in articles:
    snippets.append(article['snippet'])
snippets[:5]
```

```python
def filter_articles(articles):
    """This function iterates through the headlines, filters out the duplicates 
    and returns a list of the filtered headlines"""
    filtered_ids = []
    filtered_articles = []
    for a in articles:
        if not a['_id'] in filtered_ids:
            filtered_ids.append(a['_id'])
            filtered_articles.append(a)      
    return filtered_articles 
```

```python
filtered_articles = filter_articles(articles)
```

```python
# Out of the 910 articles that where collected, 291 were unique, the rest were duplicates
len(filtered_articles)
```

### Sentiment analysis

```python
# The snippets from the unique articles after removing the duplicates
snippets = []
for article in filtered_articles:
    snippets.append(article['snippet'])
snippets[:5]
```

```python
def pos_neg(text):
    """
    Slightly modified sentiment classification using Liu and Hu opinion lexicon.
    This function counts the number of positive, negative and neutral words
    in the tweet and classifies it depending on which polarity is more represented.
    Words that do not appear in the lexicon are considered as neutral.
    """
    from nltk.corpus import opinion_lexicon
    from nltk.tokenize import treebank

    tokenizer = treebank.TreebankWordTokenizer()
    pos_words = 0
    neg_words = 0
    tokenized_sent = [word.lower() for word in tokenizer.tokenize(text)]

    for word in tokenized_sent:
        if word in opinion_lexicon.positive():
            pos_words += 1
        elif word in opinion_lexicon.negative():
            neg_words += 1

    if pos_words > neg_words:
        return 'positive'
    elif pos_words < neg_words:
        return 'negative'
    else:
        return 'neutral'
```

```python
texts_sentiments = []    
npos = 0
nneg = 0

for text in snippets:
    sentiment = pos_neg(text)
    texts_sentiments.append([text, sentiment])
    if sentiment == 'positive':
        npos += 1
    elif sentiment == 'negative':
        nneg += 1
```

```python
def pos_neg(text):
    """
    Slightly modified sentiment classification using Liu and Hu opinion lexicon.
    This function counts the number of positive, negative and neutral words
    in the tweet and classifies it depending on which polarity is more represented.
    Words that do not appear in the lexicon are considered as neutral.
    """
    from nltk.corpus import opinion_lexicon
    from nltk.tokenize import treebank

    tokenizer = treebank.TreebankWordTokenizer()
    pos_words = 0
    neg_words = 0
    tokenized_sent = [word.lower() for word in tokenizer.tokenize(text)]

    for word in tokenized_sent:
        if word in opinion_lexicon.positive():
            pos_words += 1
        elif word in opinion_lexicon.negative():
            neg_words += 1

    if pos_words > neg_words:
        return 'positive'
    elif pos_words < neg_words:
        return 'negative'
    else:
        return 'neutral'
```

** Check how many were accurately classified using a random sample of 20 articles ** 

```python
import random
list_of_random_items = random.sample(texts_sentiments, 20)
list_of_random_items
```

```python
real_sentiments = ['negative', 'positive', 'positive', 'positive', 'negative', 
                  'negative', 'positive', 'neutral', 'neutral', 'positive', 
                  'neutral', 'neutral', 'negative', 'neutral', 'positive',
                  'positive', 'neutral', 'neutral', 'neutral', 'positive']
```

```python
sample = []
for i in range(0,20): 
    text = list_of_random_items[i][0]
    predicted_sentiment = list_of_random_items[i][1]
    real_sentiment = real_sentiments[i]
    if predicted_sentiment == real_sentiment:
        prediction = 'correct'
    else:
        prediction = 0
    sample.append([text,predicted_sentiment,real_sentiment,prediction])
sample_df = pd.DataFrame(sample, columns = ['text','predicted_sentiment','real_sentiment', 'prediction'])
sample_df   
```

```python
# Out of the 20 snippets in the sample, only 4 were misclassified
print('accuracy =', int(round(16/20*100)), '%')
```
